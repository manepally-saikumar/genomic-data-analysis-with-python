### How conserved are the sequences in the early and late outbreak?

For these calculations, we are going to use another bioinformatics package called scikit-bio.

We will be converting the clustal files into the skibio.alignment.TabularMSA class. 

What this does is creates an object that properly formats our alignment into tabbed data so we can mess around with it.

---

First let's convert the alignment file into a fasta file using AlignIO's convert function.

We do this because we can now index (or label) from the first lines of our fasta file:

```python
# to make tabular MSA, must convert to fasta file

from Bio import AlignIO

early_fasta_aln = AlignIO.convert('data/early_outbreak_GPgenes.clustal', 'clustal', 'data/early_outbreak_GPgenes.fasta', 'fasta')
early_fasta_aln
```

```python
! cat data/early_outbreak_GPgenes.fasta
```

```python
from skbio import DNA, TabularMSA

#this creates a `TabularMSA` object using our newly converted fasta file, and specifying that it is DNA
early_tabbed_alignment = TabularMSA.read('data/early_outbreak_GPgenes.fasta', format='fasta', constructor=DNA)

# reassigns the default ids (1 to n) to the fasta ID's instead of default 0,1,3 numbering
early_tabbed_alignment.reassign_index(minter='id')
early_tabbed_alignment
```

```python
type(early_tabbed_alignment[0])
```

```python
dir(early_tabbed_alignment)
```

```python
early_consensus = early_tabbed_alignment.consensus()
early_consensus
```

```python
early_conserved = early_tabbed_alignment.conservation(degenerate_mode='nan', gap_mode = 'include')
early_conserved
```

```python
early_total_seqs = (len(early_conserved))
early_total_seqs
```

```python
early_total_conserved = (early_conserved == 1).sum()
early_total_conserved
```

```python
# how conserved are these sequences from one another?

(early_total_conserved/early_total_seqs)
```

## Using matplotlib to visualize conservation across the gene

```python
import pandas as pd
import numpy as np

consensus_arr = np.array(list(str(early_consensus)))
df = pd.DataFrame(data=consensus_arr, columns=['consensus base'])
df
```

```python
df.insert(1, 'conservation score', early_conserved)
df
```

```python
base_no = np.arange(1,2407)

df.insert(0, 'base pos', base_no)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

start = 0  
stop = 2406

x_range = df['base pos'][start:stop]
y_range = df['conservation score'][start:stop]

sns.set(style='whitegrid')

plt.figure(figsize=(20,10)) 
plot = sns.lineplot(x=x_range, y=y_range, data=df)
plot.set_title("Positional Conservation from reside " + str(start) + " to " + str(stop))
```

```python
import plotly.express as px
```

```python
fig = px.bar(df, x='base pos', y='conservation score',)
fig.show()
```

## How different are the early outbreak sequences to other Ebola outbreaks?

I have another file with sequences from past outbreaks.

```python
from skbio import DNA, TabularMSA

#this creates a `TabularMSA` object using our newly converted fasta file, and specifying that it is DNA
tabbed_alignment = TabularMSA.read('data/past_Ebola_outbreaks_aln.fasta', format='fasta', constructor=DNA)

# reassigns the default ids (1 to n) to the fasta ID's instead of default 0,1,3 numbering
tabbed_alignment.reassign_index(minter='id')
tabbed_alignment
```

```python
from skbio import DistanceMatrix
from skbio.sequence.distance import hamming

dm = DistanceMatrix.from_iterable(tabbed_alignment, metric=hamming, keys=tabbed_alignment.index)
print(dm)
```

```python
plot = dm.plot(cmap = 'plasma')
```
